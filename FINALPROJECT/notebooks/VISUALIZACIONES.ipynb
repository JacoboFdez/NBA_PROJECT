```python
# Calcular la cantidad de jugadores NBA y no NBA
nba_count = df[df['NBA'] == 'Sí'].shape[0]
no_nba_count = df[df['NBA'] == 'No'].shape[0]

plt.bar(['Sí', 'No'], [nba_count, no_nba_count])
plt.title('ENTRADAS EN NBA DESDE NCAA')
plt.xlabel('NBA')
plt.ylabel('Cantidad de jugadores')

for i, v in enumerate([nba_count, no_nba_count]):
    plt.text(i, v, str(v), ha='center', va='bottom', color='black')

plt.show()
```

```python
nba_count = df[df['NBA'] == 'Sí'].shape[0]
no_nba_count = df[df['NBA'] == 'No'].shape[0]


fig, ax = plt.subplots(figsize=(6, 6))  
labels = ['Sí', 'No']
sizes = [nba_count, no_nba_count]
colors = ['#FFC154', '#3D3D3D']  
ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Personalizar título
ax.set_title('Proporción de jugadores que llegan a la NBA', fontsize=10)

# Mostrar el gráfico de pastel
plt.show()
```

```python


# Crear el subplot con 1 fila y 2 columnas
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Graficar el histograma en el primer subplot
sns.histplot(df.PPG, ax=axes[0], facecolor='dodgerblue', edgecolor='black', bins=20, color='black',kde=True)
axes[0].set_title('PTS NCAA')

# Graficar el histograma con la línea de densidad en el segundo subplot
sns.histplot(df2.PPG, ax=axes[1], facecolor='dodgerblue', edgecolor='black', bins=20, color='black', kde=True)
axes[1].set_title('PTS NBA')

# Ajustar los espacios entre subplots
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Crear el gráfico de caja y bigotes en que compararemos las características sobre los puntos anotados de los jugadores que han entrado y cuales no.
#Podemos observar que a partir de los aproximadamente 8 puntos de media por partido, los jugadores tienen muchas mas posibilidades de ir a la NBA.

plt.boxplot([df[df['NBA'] == 'Sí']['PPG'], df[df['NBA'] == 'No']['PPG']],
            labels=['NBA', 'No NBA'])

plt.xlabel('Estado de la NBA')
plt.ylabel('PPG')
plt.title('Comparación de características entre jugadores NBA y no NBA')

plt.show()
```

```python
# Crear un gráfico de barras con Seaborn
fig, ax = plt.subplots(figsize=(10, 8)) 
sns.barplot(x='NBA', y='RPG', data=df, ax=ax)  
ax.set_xlabel('Estado de la NBA', fontsize=12)
ax.set_ylabel('RPG', fontsize=12)
ax.set_title('Comparación de RPG entre jugadores NBA y no NBA', fontsize=14)

# Mostrar el gráfico de barras
plt.show()

#Una vez más observamos que hay una diferencia significativa entre los jugadores que ingresan en la NBA y los que no, en referencia a sus rebotes por partido.




```
