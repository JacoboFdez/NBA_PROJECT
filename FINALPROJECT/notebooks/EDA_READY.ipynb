```python
NCA_GROUP = NCAA.copy()
```

```python
grouped_data = NCA_GROUP.groupby('Player').mean()
```

```python
grouped_data.reset_index(inplace=True)
```

```python
grouped_data['NBA'] = grouped_data['Player'].isin(NBA['Player']).apply(lambda x: 'Sí' if x else 'No')
```

```python
grouped_data.head()
```

```python
grouped_data =  grouped_data.drop(["FGM","FGA","3PM","3PA","FTM","FTA"],axis=1)
```
