```python
import pandas as pd 
import numpy as np
```

```python
df_NBA = pd.read_csv('../data/nba_noclean.csv')
```

```python
df_NBA.head(2)
```

```python
df_NBA = df_NBA.drop(["0"],axis=1)
```

```python
df_NBA['Player'] = df_NBA["1"] + ' ' + df_NBA["2"]
```

```python
df_NBA.head(2)
```

```python
cols = list(df_NBA.columns)
cols = [cols[-1]] + cols[:-1]
df_NBA = df_NBA[cols]
```

```python
df_NBA.head(2)
```

```python
df_NBA = df_NBA.drop(['1', '2'], axis=1)
```

```python
df_NBA = df_NBA.drop(['24','25'],axis=1)
```

```python
new_columns = ["Player", "Team", "GP", "MPG", "PPG", "FGM", "FGA", "FG%", "3PM",
               "3PA", "3P", "FTM", "FTA", "FT%", "ORB", "DRB", "RPG", "APG",
               "SPG", "BPG", "TOV", "PF","Season"]
```

```python
def rename_columns(df, new_columns):
    """
    Esta función toma un dataframe y una lista de nuevas etiquetas de columna
    y devuelve el dataframe con las nuevas etiquetas de columna aplicadas.
    """
    df = df.rename(columns=dict(zip(df.columns, new_columns)))
    return df
```

```python
df_NBA = rename_columns(df_NBA, new_columns)
```
