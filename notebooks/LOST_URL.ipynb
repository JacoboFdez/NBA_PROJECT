```python
url_list = ["https://basketball.realgm.com/nba/stats/2005/Averages/All/points/All/asc/3","https://basketball.realgm.com/nba/stats/2007/Averages/All/points/All/asc/3",
"https://basketball.realgm.com/nba/stats/2010/Averages/All/points/All/asc/1","https://basketball.realgm.com/nba/stats/2013/Averages/All/points/All/asc/2","https://basketball.realgm.com/nba/stats/2015/Averages/All/points/All/asc/3",
"https://basketball.realgm.com/nba/stats/2016/Averages/All/points/All/asc/3","https://basketball.realgm.com/nba/stats/2017/Averages/All/points/All/asc/2"]
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
result_list = []

for i in url_list: 
    driver.get(i)
    try:
        driver.implicitly_wait(2)
        driver.find_element("css selector", "body > div.a__sc-np32r2-0.cqFTTR > div.Card-sc-1s2p2gv-0.a__sc-3vtlsk-0.kDNyTh.jXixFa > div.Card__CardHeader-sc-1s2p2gv-1.a__sc-3vtlsk-1.fZGtpv.cVIXeq > div > button").click()
        sleep(2)
        resultado =   driver.find_element(By.XPATH, "/html/body/div[2]/div[3]/div/table").text        
        result_list.append(resultado)
    except: 
        driver.implicitly_wait(2)
        resultado =   driver.find_element(By.XPATH, "/html/body/div[2]/div[3]/div/table").text        
        result_list.append(resultado)
driver.quit()
```

```python
def rename_columns(df, new_columns):
    """
    Esta función toma un dataframe y una lista de nuevas etiquetas de columna
    y devuelve el dataframe con las nuevas etiquetas de columna aplicadas.
    """
    df = df.rename(columns=dict(zip(df.columns, new_columns)))
    return df
```
